##DATA PREPROCESSING FOR CHINA SCHOLARSHIP DATASET

Importing Libaries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

Importing Dataset

```python
dataset = pd.read_csv('cleaned.csv')
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.describe(include = 'all')
```

```python
x = dataset.drop(['schol_id','Major','Level','Language','Tuition fees to pay'],axis = 1)
y = dataset['Tuition fees to pay']
```

Taking Care of Missing Values

```python
x.isna().sum()
```

```python
y.isna().sum()
```

```python
missing_values = ['Tuition Covered', 'Original Tuition fee']
for col in missing_values:
    x[col] = x[col].fillna(x[col].mean())
x.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
x = x.apply(label_encoder.fit_transform)
print(x)
```

 Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = ct.fit_transform(x).toarray()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 3, random_state = 0)
```
